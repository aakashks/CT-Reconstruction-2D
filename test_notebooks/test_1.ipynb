```python
x = 5.8
y = 42

angle = 2 * np.arctan(x / (2*y))
angle
```

```python
# assuming all detectors are placed at equal distances from source

for i, sod in enumerate([15, 17, 18, 19, 19.5, 21, 23, 25, 26, 27]):
    A = CreateInterceptMatrix(
        no_of_detectors=5,
        source_to_object=sod,
        source_to_detector=42,
        size_of_object=12,
        no_of_rotations=36,
        angle_bw_detectors=angle,
        resolution=5
    ).create_intercept_matrix_from_lines()
    
    plt.subplot(2, 12, i+1)
    
    plt.title(sod)
    plt.imshow(A)
    plt.axis('off')
    
    # Save the array to a CSV file
    np.savetxt(f"matrices/A_{sod}.csv", A, delimiter=",")
```
